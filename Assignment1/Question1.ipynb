** Question 1:**

* Under package Gutenberg, see what files are in it.
* Find out how many words and how many unique words in file 'austen-persuasion.txt'.
* Calculate the frequency for each words and show the top 10 most frequent words and their frequency.
* Try NOT to use FreqDist module.
* Keep the punctuation and stop words in this question.

################################################################################################################################

```python
# Import nltk
import nltk

# From nltk.corpus import gutenberg
from nltk.corpus import gutenberg
gutenberg.fileids()
```

```python
# Get all the words in words_ap
persuasion_words = nltk.corpus.gutenberg.words('austen-persuasion.txt')

# Dispaly total number of words in austen-persuasion.txt
print("Total words in austen-persuasion.txt are:", len(persuasion_words))
```

```python
# Store unique word in a unique_words
unique_words=set(persuasion_words)

# Display number of unique word in austen-persuasion.txt
print("Unique words in austen-persuasion.txt are:", len(unique_words))
```

```python
# Declare an empty dictionary 
word_count_dict={}

# Count the frequency for each word from words_ap
for word in persuasion_words:
    if word not in word_count_dict:
        word_count_dict[word] = 1
    else:
        word_count_dict[word] += 1
        
# Dictionary with all word and their frequency
print(word_count_dict)
```

################################################################################################################################
#                                             First way of doing it                                                      #
################################################################################################################################

```python
# Convert to a list for sorting.
word_count_list = []
for w in word_count_dict:
    word_count_list.append({"word":w, "count":word_count_dict[w]})

print(word_count_list)
```

** Answer **

```python
# Sort the list of words
sorted_word_list = sorted(word_count_list, key=lambda item : item["count"], reverse=True)

print(sorted_word_list[0:10])

# Prettier printing
print("\nTop 10 most frequent words from austen-persuasion.txt are:")

print("======================================================================")
print("Word",4*"\t","Frequency")
print("======================================================================")
for item in sorted_word_list[0:10]:    
    print(item["word"],4*"\t", str(item["count"]))
```

```python
# Import OrderedDict from collections
from collections import OrderedDict
# Import iterpools
import itertools

# Dictionary sorted by value
sort_wordcount=OrderedDict(sorted(word_count_dict.items(), key=lambda t: t[1], reverse=True))

# Select top 10 frequency words
x = itertools.islice(sort_wordcount.items(), 0, 10)

# Display top 10 frequency words and their corresponding frequency
print("Top 10 most frequent words from austen-persuasion.txt are:")

print("======================================================================")
print("Word",4*"\t","Frequency")
print("======================================================================")
for key, value in x:
    print ( key, 4*"\t",value)
```
