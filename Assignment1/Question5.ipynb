```python
# Import nltk
import nltk

# Import gutenberg from nltk.corpus
from nltk.corpus import gutenberg

# Import wex_text from nltk.corpus
from nltk.corpus import webtext
```

```python
# List of all files from gutenberg
gutenberg.fileids()
```

```python
# Import stopwords from nltk.corpus
from nltk.corpus import stopwords

# Get just the Shakespeare files from the Gutenberg collection
shakespeare_files = set([f for f in gutenberg.fileids() if f.startswith('shakespeare')])
# Display all shakespeare flies
print(shakespeare_files)

shakespeare_words= nltk.corpus.gutenberg.words(shakespeare_files)

# Display all shakespeare words
print(shakespeare_words)
```

```python
# Only consider words that consist entirely of letters, i.e. remove punctuation
# Lowercase them as we go.
shakespeare_letter_words = [w.lower() for w in shakespeare_words if w.isalpha()]
print(shakespeare_letter_words[0:200])
```

```python
# Import stopwords from nltk.corpus
from nltk.corpus import stopwords

# Get the English stop words file
stopwords_files = set(["english"])

# We leave the stopwords alone here
stopwords_all_words = stopwords.words(stopwords_files)
print(stopwords_all_words)
```

```python
# Create a function, since we'll need to reuse it for the modern words
def get_words(word_list):
    result = {}
    for w in word_list:
        # Remove "words" that are caused by splitting contractions.
        if not w in stopwords_all_words: #list(['d','m','s','t','th']):
            if w in result:
                result[w] = result[w] + 1
            else:
                result[w] = 1
    return result

# Pass Shakespeare letter words to get_words() function to get shakespeare updated word list
shakespeare_word_updated = get_words(shakespeare_letter_words)

# Display all shapespeare words without punctuationa and stopwrods
print("\nRemoving stopwords and punctuation the total words in shakespeare are: ", len(shakespeare_word_updated))
print("========================================================================================================================")
print(shakespeare_word_updated)
```

**Answer**

```python
# Sort the words by their frequency, descending
# Another function so it can be reused
def sort_word_counts(word_counts):
    sorted_words = sorted(word_counts, key=lambda item : word_counts[item], reverse=True)
    
    result = []
    for w in sorted_words:
        result.append({'word':w, 'count':word_counts[w]})

    return result

sorted_shakespeare_list = sort_word_counts(shakespeare_word_updated)
shakespeare_top50_words = sorted_shakespeare_list[0:50]

shakespeare_top50_words_list = [item["word"] for item in shakespeare_top50_words]

# Display top 50 shakespeare words
print("\nTop 50 Shakespeare words are:")
print("========================================================================================================================")
print(shakespeare_top50_words_list)
```
